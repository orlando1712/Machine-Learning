```python
import pandas as pd       
import numpy as np        
import matplotlib.pyplot as plt
import matplotlib as mpl


from sklearn import linear_model  
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA 
```

```python
credit = pd.read_excel("credit.xlsx")
credit.head(3)
```

```python
credit.drop(['Unnamed: 0'], axis=1,inplace=True)
credit.head(3)
```

```python
credit.columns
```

```python
credit.rename(columns={'Comp_Cliente': 'C'}, inplace=True)
credit
```

```python
carac = ['Monto_Credito', 'Edad', 'ER_Sep05', 'ER_Ago05', 'ER_Jul05', 'ER_Jun05', 'ER_May05', 'ER_Abr05', 
         'MEF_Sep05', 'MEF_Ago05', 'MEF_Jul05', 'MEF_Jun05', 'MEF_May05', 'MEF_Abr05', 'MPA_Sep05', 'MPA_Ago05', 
         'MPA_Jul05', 'MPA_Jun05', 'MPA_May05', 'MPA_Abr05']

x_scaled = StandardScaler().fit_transform(credit[carac])

print(x_scaled)
```

```python
pca = PCA(n_components=2)

pComp = pca.fit_transform(x_scaled)

credit1 = pd.DataFrame(pComp, columns=['comp1', 'comp2'])


print(credit1.head(10))
```

```python
dff = pd.concat([credit1, credit[['C']]], axis = 1)
dff.head(10)
```

```python
import matplotlib.pyplot as plt

plt.scatter(credit1['comp1'], credit1['comp2'], cmap='viridis', alpha=0.5)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Reducción PCA (2 Componentes Principales)')
plt.colorbar(label='default')
plt.show()
```

```python
plt.hist([credit1['comp1'], credit1['comp2']], color=['blue', 'orange'], alpha=0.7, bins=20, stacked=True)
plt.xlabel('Componentes Principales')
plt.ylabel('Frecuencia')
plt.title('Histogramas de Componentes Principales')
plt.legend(['Componente 1', 'Componente 2'])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(dff['comp1'], dff['C'], color='blue')
plt.xlabel("Componente Principal 1")
plt.ylabel("Y")
plt.title("Gráfico de Dispersión entre Componente Principal 1 y Y")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(dff['comp2'], dff['C'], color='blue')
plt.xlabel("Componente Principal 1")
plt.ylabel("Y")
plt.title("Gráfico de Dispersión entre Componente Principal 2 y Y")
plt.show()
```
